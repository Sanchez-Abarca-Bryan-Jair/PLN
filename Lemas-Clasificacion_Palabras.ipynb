# Tokenizacion

```python
def tokenize_v4(text):
    #delimiters = string.whitespace + string.punctuation
    acentos_enye = set([
        193, 201, 205, 211, 218,  # √Å, √â, √ç, √ì, √ö
        225, 233, 237, 243, 250,  # √°, √©, √≠, √≥, √∫
        209, 241,                 # √ë, √±
        220, 252                  # √ú, √º
    ])

    list_Mm = {193:225,#32
              201:233,#32
              205:237,#32
              211:243,#32
              218:250,#32
              209:241,#32
              220:252} #32

    set_numbers = set(range(48, 58))  # N√∫meros del 0 al 9
    set_abcdario = set(range(65, 91)).union(set(range(97, 123))).union(acentos_enye)  # Letras A-Z, a-z y letras con acentos y √±


    token = ""
    tokens = [None]*len(text)
    count = 0

    #si la posion del texto es un delimitadir y es la ultima posicion
    #se sustituye
    if ord(text[-1]) not in set_abcdario or ord(text[-1]) not in set_numbers:
        text += '.'
        n = len(text)

    for i,char in enumerate(text):
        code = ord(char)

        # Si es letra may√∫scula (A-Z), convierto a min√∫scula
        if 65 <= code <= 90:
            char = chr(code + 32)
        if code in list_Mm:
            char = chr(list_Mm[code])

        #si el caracter es una letra o numero entonces se almacena en token
        if(ord(char) in set_abcdario or ord(char) in set_numbers):
            token += char
            continue


        if token != "":
            # Limpieza: si hay letras, quitamos d√≠gitos
            has_letter = False
            for c in token:
                if ord(c) in set_abcdario:
                    has_letter = True
                    break

            if has_letter:
                new_token = ""
                for c in token:
                    if ord(c) in set_abcdario:
                        new_token += c
                token = new_token
            # Si no ten√≠a letras, es un n√∫mero puro y se queda como est√°

            if token != "":
                if token not in set_stopwords:
                    tokens[count] = token
                    count += 1
            token = ""

    return tokens[0:count]
```

```python
def reglas(palabra):
      ## verbos terminandos en ando
      n = len(palabra)
      if n > 4 and palabra[n-4:] == "ando":
          return palabra[:n-4] + "ar"

      if n > 3 and palabra[n-3:] == "aba":
          return palabra[:n-3] + "ar"

      elif n > 5 and palabra[n-5:] == "iendo":
          return palabra[:n-5] + "er"

      if n > 4 and palabra[n-4:] == "amos":
          return palabra[:n-4] + "ar"
      if n > 4 and palabra[n-4:] == "emos":
          return palabra[:n-4] + "er"
      if n > 4 and palabra[n-4:] == "imos":
          return palabra[:n-4] + "ir"

      if n > 4 and (palabra[n-3:] == "ado" or palabra[n-3:] == "ada"):
        return palabra[:n-3] + "ar"
      if n > 5 and (palabra[n-4:] == "ados" or palabra[n-4:] == "adas"):
        return palabra[:n-4] + "ar"

      if n > 4 and (palabra[n-3:] == "ido" or palabra[n-3:] == "ida"):
        return palabra[:n-3] + "er"
      if n > 5 and (palabra[n-4:] == "idos" or palabra[n-4:] == "idas"):
        return palabra[:n-4] + "er"


      if n > 4 and palabra[n-3:] == "ces":
          return palabra[:n-3] + "z"
      #if n > 4 and palabra[n-3:] == "les":
      # return palabra[:n-3]

      if n > 4 and palabra[n-4:] == "ones":
        return palabra[:n-4] + "√≥n"
      if n > 3 and palabra[n-2:] == "es" and palabra[n-3] not in VOCALS:
          return palabra[:n-2]

      if (n > 2 and palabra[n-1:] == "s" and
        palabra[n-2] in VOCALS and
        not (palabra[n-4:] == "amos" or palabra[n-4:] == "emos" or palabra[n-4:] == "imos"
             or palabra[n-2:] == "√≠s" or palabra[n-2:] == "√∫s")):
        return palabra[:n-1]

      if n > 3 and palabra[n-2:] == "as" and palabra[n-3] in VOCALS:
          return palabra[:n-1]
      if n > 3 and palabra[n-2:] == "os" and palabra[n-3] in VOCALS:
          return palabra[:n-1]

      if n > 7 and palabra[n-5:] == "mente":
        return palabra[:n-5]

      if n > 2 and palabra[n-1:] == "√≥":
        return palabra[:n-1] + "ar"

      if n > 5 and palabra[n-4:] == "aron":
        return palabra[:n-4] + "ar"

      return palabra
```

```python
def lematizacion_reglas_excepciones(palabra):
    if palabra in lemas_excepcciones:
        return lemas_excepcciones[palabra]
    return reglas(palabra)
```

```python
def lematizar(texto): #Recibe documento tokenizado
    resultado = [None]*len(texto)
    for i in range(len(texto)):
        lema = lematizacion_reglas_excepciones(texto[i])
        if lema is None:
            lema = tokens[i]
        resultado[i] = lema
    return resultado
```

## Con similitud coseno

```python
import numpy as np

def indices_matriz(vocabulario, palabra):
  for j in range(len(vocabulario)):
    if palabra == vocabulario[j]:
      return j
  return None

def matriz_concurrencia(texto_tokenizado, vocabulario):
    matriz_concurrencia = np.zeros((len(vocabulario), len(vocabulario)))

    for i in range(len(texto_tokenizado) - 1):
        palabra = texto_tokenizado[i]
        palabra_que_sigue = texto_tokenizado[i+1]

        indice_palabra_actual = indices_matriz(vocabulario, palabra)
        indice_palabra_que_sigue = indices_matriz(vocabulario, palabra_que_sigue)

        if indice_palabra_actual is None or indice_palabra_que_sigue is None:
            continue

        if indice_palabra_actual == indice_palabra_que_sigue:
            matriz_concurrencia[indice_palabra_actual][indice_palabra_que_sigue] += 1
        else:
            matriz_concurrencia[indice_palabra_actual][indice_palabra_que_sigue] += 1
            matriz_concurrencia[indice_palabra_que_sigue][indice_palabra_actual] += 1
    return matriz_concurrencia

def similitud_coseno(palabra1, palabra2, matriz_concurrencia, vocabulario):
    indice_palabra1 = indices_matriz(vocabulario, palabra1)
    indice_palabra2 = indices_matriz(vocabulario, palabra2)

    #Si alguna palabra no est√° en el vocabulario, su similitud es 0
    if indice_palabra1 is None or indice_palabra2 is None:
        return 0.0

    vec_palabra1 = matriz_concurrencia[indice_palabra1]
    vec_palabra2 = matriz_concurrencia[indice_palabra2]

    producto_punto = np.dot(vec_palabra1, vec_palabra2)
    norma_palabra1 = np.linalg.norm(vec_palabra1)
    norma_palabra2 = np.linalg.norm(vec_palabra2)

    #Manejo de divisi√≥n por cero
    if norma_palabra1 == 0 or norma_palabra2 == 0:
        return 0.0 #Si una palabra no tiene vecinos, su similitud es 0
    else:
        cos_sim = producto_punto / (norma_palabra1 * norma_palabra2)
        return cos_sim
```

```python
def clasificar_palabra_por_coseno(palabra_usuario, corpus_lemas_dict, mapa_temas):
    palabra_usuario = lematizacion_reglas_excepciones(palabra_usuario)
    print(f"Palabra: '{palabra_usuario}'")

    puntajes_finales = {}

    #Iterar por los 12 documentos lematizados
    for doc_key, palabras_del_documento in corpus_lemas_dict.items():
        vocabulario_doc = PalabrasUnicas(palabras_del_documento)

        #Si la palabra del usuario no est√° en este documento, no podemos calcular su vector y su puntaje es 0
        if palabra_usuario not in vocabulario_doc:
            puntajes_finales[doc_key] = 0.0
            continue

        matriz_doc = matriz_concurrencia(palabras_del_documento, vocabulario_doc)

        similitudes_doc = []
        for lema_doc in vocabulario_doc:
            sim = similitud_coseno(palabra_usuario, lema_doc, matriz_doc, vocabulario_doc)
            similitudes_doc.append(sim)

        #Tomar los 5 m√°s altos y sumarlos
        similitudes_doc.sort(reverse=True)
        top_5_similitudes = similitudes_doc[:5]
        suma_top_5 = sum(top_5_similitudes)

        #print(f"    Suma para {doc_key}: {suma_top_5:.4f}")
        puntajes_finales[doc_key] = suma_top_5

    for doc, puntaje in puntajes_finales.items():
        print(f"  {doc} ({mapa_temas[doc]}): {puntaje:.4f}")

    doc_ganador = max(puntajes_finales, key=puntajes_finales.get)
    puntaje_ganador = puntajes_finales[doc_ganador]
    tema_ganador = mapa_temas[doc_ganador]

    print(f"La palabra '{palabra_usuario}' (lema: '{palabra_usuario}') Se relaciona m√°s con:")
    print(f"Documento: {doc_ganador}")
    print(f"Tema: {tema_ganador}")
    print(f"Puntaje: {puntaje_ganador:.4f}")
```

```python
palabra_ingresada_3 = input("Introduce una palabra: ")

clasificar_palabra_por_coseno(palabra_ingresada_3, corpus_completo_lematizado, mapa_temas_completo)
```

```python
from nltk.corpus import wordnet as wn

def clasificar_palabra_por_wordnet(palabra_usuario, corpus_lemas_dict, mapa_temas):
    palabra_usuario = lematizacion_reglas_excepciones(palabra_usuario)
    print(f"Palabra: '{palabra_usuario}'")

    #Obtener el synset de la palabra
    try:
        syn_usuario = wn.synsets(palabra_usuario, lang='spa')[0]
        print(f"Significado de '{palabra_usuario}': {syn_usuario.name()} ({syn_usuario.definition()})")
    except IndexError:
        print(f"La palabra no se encontr√≥.")
        return

    puntajes_finales = {}

    for doc_key, palabras_del_documento in corpus_lemas_dict.items():
        similitudes_doc = []

        for lema_doc in palabras_del_documento:

            similitud = 0.0
            try:
                #Aqui se busca el synset pero en el documento
                syn_doc = wn.synsets(lema_doc, lang='spa')[0]

                #Calcular la similitud sem√°ntica
                sim = syn_usuario.path_similarity(syn_doc)

                # Si no hay un camino, devuelve None
                if sim is None:
                    sim = 0.0
            except IndexError:
                #La palabra del documento no se encontr√≥ en WordNet
                sim = 0.0

            similitudes_doc.append(sim)

        similitudes_doc.sort(reverse=True)
        top_5_similitudes = similitudes_doc[:5]
        suma_top_5 = sum(top_5_similitudes)

        puntajes_finales[doc_key] = suma_top_5

    for doc, puntaje in puntajes_finales.items():
        print(f"  {doc} ({mapa_temas[doc]}): {puntaje:.4f}")

    doc_ganador = max(puntajes_finales, key=puntajes_finales.get)
    puntaje_ganador = puntajes_finales[doc_ganador]
    tema_ganador = mapa_temas[doc_ganador]

    print(f"La palabra '{palabra_usuario}' (lema: '{palabra_usuario}') se relaciona m√°s con:")
    print(f"Documento: {doc_ganador}")
    print(f"Tema: {tema_ganador}")
    print(f"Puntaje: {puntaje_ganador:.4f}")
```
