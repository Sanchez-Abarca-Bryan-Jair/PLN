# Tokenizacion

```python
def tokenize(text):
    #delimiters = string.whitespace + string.punctuation
    acentos_enye = set([
        193, 201, 205, 211, 218,  # Á, É, Í, Ó, Ú
        225, 233, 237, 243, 250,  # á, é, í, ó, ú
        209, 241,                 # Ñ, ñ
        220, 252,                  # Ü, ü
        194, 226,        # Â â
        198, 230,        # Æ æ
        199, 231,        # Ç ç
        202, 234,        # Ê ê
        203, 235,        # Ë ë
        206, 238,        # Î î
        207, 239,        # Ï ï
        212, 244,        # Ô ô
        219, 251,        # Û û
        220, 252,        # Ü ü  (ya lo tienes, pero no pasa nada)
        255,             # ÿ
        232,
        44,
        8217,
        224,
        226,
        232,
        233,
        376,             # Ÿ  (fuera de 255)
        338, 339         # Œ œ (fuera de 255)
    ])

    set_numbers = set(range(48, 58))  # Números del 0 al 9
    set_abcdario = set(range(65, 91)).union(set(range(97, 123))).union(acentos_enye)  # Letras A-Z, a-z y letras con acentos y ñ

    token = ""
    tokens = [None]*len(text)
    count = 0

    #si la posion del texto es un delimitadir y es la ultima posicion
    #se sustituye
    if ord(text[-1]) not in set_abcdario or ord(text[-1]) not in set_numbers:
        text += '.'
        n = len(text)
    for i,char in enumerate(text):
        #si el caracter es una letra entonces se almacena en token
        if(ord(char) in set_abcdario or ord(char) in set_numbers):
            token += char
            continue


        if token != "":
            # Limpieza: si hay letras, quitamos dígitos
            has_letter = False
            for c in token:
                if ord(c) in set_abcdario:
                    has_letter = True
                    break

            if has_letter:
                new_token = ""
                for c in token:
                    if ord(c) in set_abcdario:
                        new_token += c
                token = new_token
            # Si no tenía letras, es un número puro y se queda como está

            if token != "":
                  tokens[count] = token
                  count += 1
            token = ""

    return tokens[0:count]
```

# N gramas

```python
def ngCaracteres(texto, n=3):
  texto = texto.lower().replace(' ', '')
  return [texto[i:i+n] for i in range(len(texto)-n+1)]
```

```python
def ngPalabras(texto, n):
  palabras = texto.lower().split()
  return [tuple(palabras[i:i+n]) for i in range(len(palabras)-n+1)]
```

```python
def contar(lista):
  d = {}
  for x in lista:
    d[x] = d.get(x, 0) + 1
  return d
```

```python
def ngramas(texto, n=3):
  return [texto[i:i+n] for i in range(len(texto)-n+1)]
```

```python
def prob_palabra(palabra, idioma, freq_por_idioma):
    return freq_por_idioma[idioma].get(palabra, 0)
```

```python
def perfilNgramas(texto,n=3):
  perfil = {}
  for ng in ngCaracteres(texto,n):
    perfil[ng] = perfil.get(ng,0) + 1
  return perfil
```

```python
def dsitancia(p1,p2):
  dist = 0
  todos = set(p1.keys()) | set(p2.keys())
  for k in todos:
    dist += abs(p1.get(k,0) - p2.get(k,0))
  return dist
```

# Palabras clave


```python
def PalabrasClave(texto,top=5):
  uni = contar(ngPalabras(texto,1))
  bi = contar(ngPalabras(texto,2))
  tri = contar(ngPalabras(texto,3))

  uni_top = sorted(uni.items(), key=lambda x: x[1], reverse=True)[:top]
  bi_top = sorted(bi.items(), key=lambda x: x[1], reverse=True)[:top]
  tri_top = sorted(tri.items(), key=lambda x: x[1], reverse=True)[:top]

  return uni_top, bi_top, tri_top
```

# Edicion

```python
def edits1(p):
    caracteres = "abcdefghijklmnopqrstuvwxyzáéíóúüñàèìòùâêîôûçœæ'’"
    splits = [(p[:i], p[i:]) for i in range(len(p)+1)]

    deletes = [L + R[1:] for L, R in splits if R]
    inserts = [L + c + R for L, R in splits for c in caracteres]
    replaces = [L + c + R[1:] for L, R in splits for c in caracteres if R]
    transposes = [L + R[1] + R[0] + R[2:] for L, R in splits if len(R) > 1]

    return set(deletes + inserts + replaces + transposes)

def edits2(p):
    return {e2 for e1 in edits1(p) for e2 in edits1(e1)}
```

```python
def CorregirPalabra(palabra, idioma, vocab_por_idioma, freq_por_idioma):
    palabra = palabra.lower()
    vocabulario = vocab_por_idioma[idioma]
    freq = freq_por_idioma[idioma]

    if palabra in vocabulario:
        return palabra

    cands1 = [w for w in edits1(palabra) if w in vocabulario]
    if cands1:
        return max(cands1, key=lambda w: freq.get(w, 0))

    cands2 = [w for w in edits2(palabra) if w in vocabulario]
    if cands2:
        return max(cands2, key=lambda w: freq.get(w, 0))

    return palabra

```

# Idiomas

```python
modelo_idiomas = {
    'Espanish': perfilNgramas(texto_ES),
    'French': perfilNgramas(texto_FC),
    'Italian': perfilNgramas(texto_IT),
    'English': perfilNgramas(texto_EN)
}
```

```python
def similitud_coseno(p1, p2, eps=1e-12):
    dot = 0.0 # producto punto
    norm1 = 0.0
    norm2 = 0.0

    for k, v in p1.items():
        norm1 += v*v
        if k in p2:
            dot += v * p2[k]

    for v in p2.values():
        norm2 += v*v

    norm1 = math.sqrt(norm1)
    norm2 = math.sqrt(norm2)

    if norm1 < eps or norm2 < eps:
        return 0.0

    return dot / (norm1 * norm2)
```

```python
def distancia_coseno(p1, p2):
    return 1.0 - similitud_coseno(p1, p2)
```

```python
def detectar_idioma(texto):
    perfil = perfilNgramas(texto)
    distancias = {}

    for idioma, modelo in modelo_idiomas.items():
        distancias[idioma] = distancia_coseno(perfil, modelo)

    return min(distancias, key=distancias.get)
```

# Prediccion de palabras

```python
def entrenar_bigrama(tokens):
    trans = {} # diccionario de transicciones
    unigrams = contar(tokens) # frecuencia de cada palabra

    for w1, w2 in zip(tokens[:-1], tokens[1:]): # usa bigramas
        if w1 not in trans:
            trans[w1] = {}
        trans[w1][w2] = trans[w1].get(w2, 0) + 1 #suma 1 a la transicion

    return trans, unigrams
```

```python
modelo_pred = {}

for lang, toks in {
    "es": texto_es,
    "en": texto_ingles,
    "fr": texto_frances,
    "it": texto_italiano,
}.items():
    trans, uni = entrenar_bigrama(toks)
    modelo_pred[lang] = {"trans": trans, "uni": uni}

```

```python
def predecir_siguiente(palabra, idioma, modelo_pred):
    palabra = palabra.lower()
    trans = modelo_pred[idioma]["trans"] #diccionario de transicciones
    uni = modelo_pred[idioma]["uni"]# frecuencia de palabra en el corpus

    # fallback = palabra más frecuente del idioma
    fallback = max(uni, key=uni.get) if len(uni) else ""

    #garantiza que siempre devuelva una prediccion, ya que si no encentra nada
    # devuelve la palabra mas comun en el idioma ele
    if palabra not in trans or len(trans[palabra]) == 0:
        return fallback

    # devuelve la más probable después de la palabra
    return max(trans[palabra], key=trans[palabra].get)

```

```python
stop_por_idioma = {
  "en": {"the","and","of","to","in","a","is","it","that","for","on","as","with","this","are","be","or","an"},
  "es": {"la","el","los","las","y","de","en","para","que","un","una","por","con","del","al","se","ha","es"},
  "fr": {"la","le","les","et","de","des","en","un","une","du","à","au","aux","que","qui","dans"},
  "it": {"la","il","lo","i","gli","le","e","di","del","della","in","un","una","che","a","da","per"},
}
```

```python
def predecir_siguiente_sin_stop(palabra, idioma, modelo_pred, stop_por_idioma):
    palabra = palabra.lower()
    trans = modelo_pred[idioma]["trans"]
    uni = modelo_pred[idioma]["uni"]
    stop = stop_por_idioma.get(idioma, set())

    # fallback (siempre algo): palabra más frecuente que NO sea stopword
    candidatos_fb = [w for w in uni.keys() if w not in stop and len(w) > 2]
    fallback = max(candidatos_fb, key=lambda w: uni[w]) if candidatos_fb else max(uni, key=uni.get)

    if palabra not in trans or len(trans[palabra]) == 0:
        return fallback

    # ordenar siguientes por frecuencia (desc)
    siguientes = sorted(trans[palabra].items(), key=lambda x: x[1], reverse=True)

    # elegir el primero que no sea stopword
    for w2, _ in siguientes:
        if w2 not in stop and len(w2) > 2:
            return w2

    # si todos son stopwords, regresa el top1 (o fallback)
    return siguientes[0][0] if siguientes else fallback
```

```python

#funciona peor a usar bigrmas ya que no se eliminaron las stopwords,
# lo que hace que casi siempre las devuelva como resultado
def predecir_bayes(palabra, idioma, modelo_pred, alpha=1.0):
    palabra = palabra.lower()

    trans = modelo_pred[idioma]["trans"]
    uni   = modelo_pred[idioma]["uni"]

    vocab = list(uni.keys())
    V = len(vocab)

    mejor_palabra = None
    mejor_score = -float("inf")

    for w in vocab:
        # P(w)  (prior)
        Pw = (uni.get(w, 0) + alpha) / (sum(uni.values()) + alpha * V)

        # P(palabra | w)  (likelihood)
        count_w_prev = trans.get(w, {}).get(palabra, 0)
        total_w = sum(trans.get(w, {}).values())

        Pprev_given_w = (count_w_prev + alpha) / (total_w + alpha * V)

        score = math.log(Pw) + math.log(Pprev_given_w)

        if score > mejor_score:
            mejor_score = score
            mejor_palabra = w

    return mejor_palabra

```
